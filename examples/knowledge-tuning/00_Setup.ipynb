## Install local utilities

This repository includes helper utilities (for example, `ai_tools` in `src/`) that the notebooks use during development. Instead of building and installing wheel files from the example, prefer one of the following approaches:

1. Editable install from the repository root (recommended for development):

```bash
%pip install -e ../../
```

2. Or add the repository `src/` directory to `PYTHONPATH` or append it to `sys.path` inside the notebooks:

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent / 'src'))
```

Editable installs are simpler for iterative development and avoid the need to manage local wheel files inside the example folders.
